### Get distance of players to ball carrier and orientation and direction difference 

```python
import numpy as np
import pandas as pd
from tqdm import tqdm

# Define a function to calculate distance, orientation, and direction difference to the ballcarrier
def calculate_metrics(df):
    x1, y1, x2, y2 = df['x'], df['y'], df['ballcarrier_x'], df['ballcarrier_y']
    player_o, ballcarrier_o = df['o'], df['ballcarrier_o']
    player_dir, ballcarrier_dir = df['dir'], df['ballcarrier_dir']

    # Calculate the distance to the ballcarrier
    distance = np.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)

    # Calculate the absolute orientation difference relative to the ballcarrier
    angle_diff_o = np.abs(player_o - ballcarrier_o) % 360
    orientation_diff = angle_diff_o.where(angle_diff_o < (360 - angle_diff_o), 360 - angle_diff_o)

    # Calculate the absolute direction difference relative to the ballcarrier
    angle_diff_dir = np.abs(player_dir - ballcarrier_dir) % 360
    direction_diff = angle_diff_dir.where(angle_diff_dir < (360 - angle_diff_dir), 360 - angle_diff_dir)

    return pd.DataFrame({'dis_to_ballcarrier': distance,
                         'orientation_diff_to_ballcarrier': orientation_diff,
                         'direction_diff_to_ballcarrier': direction_diff})

# Calculate the metrics using NumPy vectorization
tqdm.pandas()
final_metrics = calculate_metrics(final)

# Combine the calculated metrics with the original DataFrame
final = pd.concat([final, final_metrics], axis=1)
final
```

```python
import pandas as pd
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from sklearn.model_selection import train_test_split

# Sort tackler dataframe by the sequence identifiers
tackler.sort_values(by=['gameId', 'playId', 'nflId', 'frameId'], inplace=True)

feature_columns = ['x', 'y', 's', 'a', 'o', 'dir', 'ballcarrier_o', 'ballcarrier_dir', 'ballcarrier_dis',
                   'ballcarrier_x', 'ballcarrier_y', 'ballcarrier_s', 'ballcarrier_a', 'dis_to_ballcarrier',
                   'orientation_diff_to_ballcarrier', 'direction_diff_to_ballcarrier', 'opp1Dist', 
       'opp1X', 'opp1Y', 'opp1Dir', 'opp1O', 'opp1S', 'opp1A', 'opp2Dist',
       'opp2X', 'opp2Y', 'opp2Dir', 'opp2O', 'opp2S', 'opp2A',
       'opp3Dist', 'opp3X', 'opp3Y', 'opp3Dir', 'opp3O',
       'opp3S', 'opp3A', 'opp4Dist', 'opp4X', 'opp4Y',
       'opp4Dir', 'opp4O', 'opp4S', 'opp4A', 'opp5Dist',
       'opp5X', 'opp5Y', 'opp5Dir', 'opp5O', 'opp5S', 'opp5A', 'opp6Dist',
    'opp6X', 'opp6Y', 'opp6Dir', 'opp6O', 'opp6S', 'opp6A',
       'opp7Dist',  'opp7X', 'opp7Y', 'opp7Dir', 'opp7O',
       'opp7S', 'opp7A', 'opp8Dist', 'opp8X', 'opp8Y',
       'opp8Dir', 'opp8O', 'opp8S', 'opp8A', 'opp9Dist', 
       'opp9X', 'opp9Y', 'opp9Dir', 'opp9O', 'opp9S', 'opp9A',
       'opp10Dist', 'opp10X', 'opp10Y', 'opp10Dir', 
        'opp10O', 'opp10S', 'opp10A', 'mean_dis_to_ballcarrier',
        'variance_dis_to_ballcarrier',
        'mate1Dist', 
       'mate1X', 'mate1Y', 'mate1Dir', 'mate1O', 'mate1S', 'mate1A',
       'mate2Dist', 'mate2X', 'mate2Y', 'mate2Dir', 'mate2O',
       'mate2S', 'mate2A', 'mate3Dist', 'mate3X', 'mate3Y',
       'mate3Dir', 'mate3O', 'mate3S', 'mate3A', 'mate4Dist',
       'mate4X', 'mate4Y', 'mate4Dir', 'mate4O', 'mate4S', 'mate4A',
       'mate5Dist', 'mate5X', 'mate5Y', 'mate5Dir', 'mate5O',
       'mate5S', 'mate5A', 'mate6Dist', 'mate6X', 'mate6Y',
       'mate6Dir', 'mate6O', 'mate6S', 'mate6A', 'mate7Dist',
       'mate7X', 'mate7Y', 'mate7Dir', 'mate7O', 'mate7S', 'mate7A',
       'mate8Dist', 'mate8X', 'mate8Y', 'mate8Dir', 'mate8O',
       'mate8S', 'mate8A', 'mate9Dist', 'mate9X', 'mate9Y',
       'mate9Dir', 'mate9O', 'mate9S', 'mate9A', 'mate10Dist',
       'mate10X', 'mate10Y', 'mate10Dir', 'mate10O', 'mate10S', 'mate10A']

# Group by sequence identifiers and create list of sequences
sequences = tackler.groupby(['gameId', 'playId', 'nflId'])[feature_columns + ['time_remaining']].apply(lambda x: x.values.tolist())

# Split the dataset into inputs (X) and target (y)
X = sequences.apply(lambda x: [i[:-1] for i in x]).tolist()

# In this case, we don't just take the final time_remaining value.
# Instead, we want to keep the time dimension in our target.
# So `y` remains as sequences of `time_remaining` values.
y = sequences.apply(lambda x: [i[-1] for i in x]).tolist()
# Padding sequences for consistent input size for both features and labels
max_sequence_length = max(len(seq) for seq in X)  # Find the longest sequence
X_padded = tf.keras.preprocessing.sequence.pad_sequences(X, maxlen=max_sequence_length, padding='post', dtype='float32')

# Padding the labels
y_padded = np.zeros((len(y), max_sequence_length, 1))  # Assuming that '1' is for a single regression output
for i, sequence in enumerate(y):
    for t, value in enumerate(sequence):
        y_padded[i, t, 0] = value

# Model that returns a sequence using LSTM
model = Sequential([
    LSTM(50, input_shape=(None, len(feature_columns)), return_sequences=True, dropout=0.2, recurrent_dropout=0.2),
    LSTM(50, return_sequences=True, dropout=0.2, recurrent_dropout=0.2),
    LSTM(50, return_sequences=True, dropout=0.2, recurrent_dropout=0.2),
    tf.keras.layers.TimeDistributed(Dense(1))
])

# Compile the model with the RMSE metric
model.compile(optimizer='adam', loss='mse', metrics=[tf.metrics.RootMeanSquaredError()])

# Train the model - y_padded is used here instead of y_train
history = model.fit(X_padded, y_padded, epochs=20, batch_size=32, validation_split=0.2)
```

```python
def calculate_angle_of_pursuit(row):
    """
    Calculates the angle of pursuit the tackler needs to take in order to intercept the ball carrier
    using the predicted time until contact.

    Parameters:
    - row: Row of data containing relevant information, including the predicted time until contact.

    Returns:
    - Angle of pursuit required (in degrees) for tackler to intercept the ball carrier.
    """
    # Use the predicted time until contact to calculate future positions
    predicted_time_to_intercept = row['predicted_time_until_contact']
    future_tackler_x = row['x'] + row['s'] * np.cos(np.radians(row['dir'])) * predicted_time_to_intercept
    future_tackler_y = row['y'] + row['s'] * np.sin(np.radians(row['dir'])) * predicted_time_to_intercept
    future_bc_x = row['ballcarrier_x'] + row['ballcarrier_s'] * np.cos(np.radians(row['ballcarrier_dir'])) * predicted_time_to_intercept
    future_bc_y = row['ballcarrier_y'] + row['ballcarrier_s'] * np.sin(np.radians(row['ballcarrier_dir'])) * predicted_time_to_intercept

    # Calculate angle of pursuit
    tackler_vector = np.array([np.cos(np.radians(row['dir'])), np.sin(np.radians(row['dir']))])
    to_future_bc_vector = np.array([future_bc_x - future_tackler_x, future_bc_y - future_tackler_y])
    to_bc_magnitude = np.linalg.norm(to_future_bc_vector)
    to_bc_unit_vector = to_future_bc_vector / to_bc_magnitude
    dot_product = np.dot(tackler_vector, to_bc_unit_vector)
    det = tackler_vector[0] * to_bc_unit_vector[1] - tackler_vector[1] * to_bc_unit_vector[0]
    angle = np.degrees(np.arctan2(det, dot_product))

    # Normalize angle to be between 0 and 180 degrees
    angle = (angle + 360) % 360
    if angle > 180:
        angle = 360 - angle
    return angle

tqdm.pandas()
tackler['angle_of_pursuit'] = tackler.progress_apply(lambda row: calculate_angle_of_pursuit(row), axis=1)
tackler['optimal_dir'] = (tackler['dir'] + tackler['angle_of_pursuit']) % 360
tackler
```
